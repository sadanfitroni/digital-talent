## 1. Deskripsi Data 
Data yang digunakan diperoleh dari Badan Pusat Statistik melalui laman data.go.id yaitu Data Inflasi Bulanan per Kabupaten/Kota yang ada di Indonesia. Data set inflasi ini dijabarkan per bulan untuk masing-masing wilayah dengan data mulai dari tahun 1979 - 2014. 

Berikut ini informasi mengenai dataset yang digunakan 

- Terdiri dari 6 kolom yaitu, tahun, bulan, namaa_kabkota, ihk, inflasi, latitude, dan longitude. 
- Terdiri dari 15246 rows, dengan 7 kolom. 
- Format data yaitu csv dengan ukuran file 752 Kb 

Berikut informasi mengenai masing-masing kolom pada data set

1. Tahun     : Tahun bersangkutan
2. Bulan     : Bulan bersangkutan 
3. ihk       : Indeks harga konsumen 
4. inflasi   : Kenaikan harga (dalam bentuk persen)
5. latitude  : Koordinat lintang lokasi kab/kota 
6. longitude : Koordinat bujur lokasi kab/kota 
   

```python
# Membaca data set dengan pandas 
import pandas as pd
data_inflasi = pd.read_csv('processedinflasibulanankabupatenkota19792014.csv', delimiter = ',', encoding = 'utf-8')
data_inflasi
```

## Informasi tentang type data 

```python
data_inflasi.info()
```

## 2. Statistik tentang Data 
Perhitungan statistik yang akan dihitung adalah mean, max, min, inflasi untuk setiap tahun, bulan, dan per kabupaten/kota. 

```python
jumlah_kabkota = data_inflasi['nama_kabkota'].value_counts().count()
#rata_rata = data_inflasi.groupby('tahun').mean()['inflasi']
maximum = data_inflasi[['tahun','bulan','inflasi','ihk','nama_kabkota']].max()
minimum = data_inflasi[['tahun','bulan','inflasi','ihk','nama_kabkota']].min()
rata_ihk = data_inflasi['ihk'].mean()
ihk_mean = round(rata_ihk,2)
rata_inflasi = data_inflasi['inflasi'].mean()
inflasi_mean = round(rata_inflasi,2)

print("====================================================")
print("Data Inflasi di Indonesia Berdasarkan Kabupaten/Kota")
print("====================================================")
print("Jumlah Kab/Kota :", jumlah_kabkota, "Kab/Kota")
print("====================================================")
print("Rata_rata Inflasi per Tahun :")
print("====================================================")
print("Mean Index Harga Inflasi :")
print(inflasi_mean)
print("====================================================")
print("Mean Index Harga Konsumen :")
print(ihk_mean)
print("====================================================")
print("Inflasi Tertinggi:", maximum,'%')
print("====================================================")
print("Inflasi terendah  :", minimum,'%')
print("====================================================")
```

## 3. Visualisasi Data 

```python
# Visualisasi Laju Inflasi per Tahun
ax = data_inflasi[['tahun','inflasi']].groupby(['tahun']).mean()
ax = ax.plot(kind='bar', title='Total Inflasi per tahun', legend=True, fontsize=12, figsize=(15,6))
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Laju Inflasi (%)", fontsize=6)
```

## Analisa
Berdasarkan diagram di atas, bahwa tingkat inflasi paling tinggi terjadi pada tahun 1998. Hal ini karena pada tahun tersebut terjadi krisis moneter yang disebabkan oleh neraca pembayaran. 

```python
# Inflasi perkabupaten 
# Visualisasi Laju Inflasi per Tahun
ax = data_inflasi[['nama_kabkota','inflasi']].groupby(['nama_kabkota']).mean()
ax = ax.plot(kind='bar', title='Total Inflasi per kabupaten', legend=True, fontsize=12, figsize=(15,6))
ax.set_xlabel("Kab/Kota", fontsize=10)
ax.set_ylabel("Laju Inflasi (%)", fontsize=6)
```

Dari diiagram di atas kita bisa melihat variasi dari laju inflasi untuk masing-masing kab/kota. Inflasi paling tinggi yaitu di Kab Banyuwangi, dan daerah dengan tingkat inflasi terendah yaitu Kota Tual.
